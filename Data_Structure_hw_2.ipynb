```python
import random
import time
import matplotlib.pyplot as plt

def generate_array(length):
  array = [random.randint(1,1000) for i in range(length)]
  return array

def max(lst):
  x = lst[0]
  for i in range(1,len(lst)):
    if lst[i]>x:
      x = lst[i]
  return x

array_length = [10,100,1000,100000,1000000]
array_time = []
for i in array_length:
  array = generate_array(i)
  start = time.time()
  max(array)
  end = time.time()
  execution_time = end-start
  array_time.append(execution_time)


plt.plot(array_length,array_time)
plt.title('Time Taken By Program For Different Input Sizes')
```

```python
import time
import matplotlib.pyplot as plt

def log_2(n):
  for i in range(n):
    i = 1
    while i<n:
      i *= 2

input_sizes = [10,100,1000,100000,1000000]
execution_times = []

for size in input_sizes:
  start= time.time()
  log_2(size)
  end= time.time()
  execution_time = end-start
  execution_times.append(execution_time)

plt.plot(input_sizes,execution_times)
plt.title('Time Taken By Program For Different Input Sizes')

```

```python
import time
import random
import matplotlib.pyplot as plt

def generate_array(length):
  array = [random.randint(1,1000) for i in range(length)]
  return array

def r(a):
  n = len(a)
  x = [None]*n
  for i in range(n):
    x[n-1-i] = a[i]
  return x

array_length = [10,100,1000,100000,1000000]
array_time = []
for i in array_length:
  array = generate_array(i)
  start = time.time()
  r(array)
  end = time.time()
  execution_time = end-start
  array_time.append(execution_time)


plt.plot(array_length,array_time)
plt.title('Time Taken By Program For Different Input Sizes')
```

```python
import time
import random
import matplotlib.pyplot as plt

def generate_array(length):
  array = [random.randint(1,1000) for i in range(length)]
  return array

def r(a):
  n = len(a)
  for i in range(n//2):
    a[i], a[n-1-i] = a[n-1-i], a[i]
  return a

array_length = [10,100,1000,100000,1000000]
array_time = []
for i in array_length:
  array = generate_array(i)
  start = time.time()
  r(array)
  end = time.time()
  execution_time = end-start
  array_time.append(execution_time)
```
